# Handling outliers

```python
sns.boxplot(df)
```

```python
sns.boxplot(df['fare'])
```

```python
q1,q2,q3 = np.percentile(df['fare'],[25,50,75])
print(q1,q2,q3)
```

```python
q1,q2,q3 = np.percentile(df['fare'],[25,50,75])
print(q1,q2,q3)
IQR = q3-q1
print(IQR)
lower = q1-1.5*IQR
higher = q3+1.5*IQR
print(lower,higher)

outlier=[]
for i in df['fare']:
    if i<lower or i>higher:
        outlier.append
print(outlier)
```

```python
np.where(df['fare']>higher)
```

```python
np.where(df['fare']>higher,higher,df['fare'])
```

```python
df['fare']=np.where(df['fare']>higher,higher,df['fare'])
```

```python
#capping replace with higher/ flooring replace with lower
```

```python
sns.boxplot(df['fare'])
```

```python
q1,q2,q3 = np.percentile(df['age'],[25,50,75])
print(q1,q2,q3)
IQR = q3-q1
print(IQR)
lower_limit = q1 - 1.5*IQR
upper_limit = q3 + 1.5*IQR
print(lower_limit,upper_limit)

outlier =[]
for i in df['fare']:
  if i<lower_limit or i>upper_limit:
    outlier.append(i)
print(outlier)
```

```python
df['age']=np.where(df['age']>upper_limit,upper_limit,df['age'])
df['age']=np.where(df['age']<lower_limit,lower_limit,df['age'])
```

```python
import numpy as np

# Make a copy to avoid altering original
X = x.copy()

# Convert boolean columns to integers
for col in X.select_dtypes(include=['bool']).columns:
    X[col] = X[col].astype(int)

# Replace infinities and NaN with 0
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(0)

# Ensure all values are non-negative (chi2 requires this)
X = np.abs(X)
```

```python
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif, f_classif

# select top 2 features using chi2
#selector = SelectKBest(chi2, k=2)
#X_new = selector.fit_transform(X, y)
#print(X_new)

# # select top 2 features using mutual_info_classif
selector = SelectKBest(mutual_info_classif, k=2)
X_new = selector.fit_transform(X, y)
print(X_new)

# # select top 2 features using f_classif
# selector = SelectKBest(f_classif, k=2)
# X_new = selector.fit_transform(X, y)
# print(X_new)
```

```python
selected_features = x.columns[selector.get_support()]
print("Selected Features:", selected_features.tolist())
```
